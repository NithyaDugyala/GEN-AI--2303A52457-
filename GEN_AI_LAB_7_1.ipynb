```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
file_path = "/content/diabetesgn.csv"
df = pd.read_csv(file_path)
df.head()
```

```python
X = df.drop(columns=['Outcome'])
y = df['Outcome']
X
```

```python
y
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
model = Sequential([
    Dense(8, activation='relu', input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(16, activation='relu'),
    Dense(20, activation='relu'),
    Dense(10, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

```python
model.compile(optimizer='adadelta', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=150, batch_size=64, validation_data=(X_test, y_test))
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
model.save("diabetes_model.h5")
```

```python
def predict_diabetes(input_data):
    input_scaled = scaler.transform([input_data])
    prediction = model.predict(input_scaled)
    return prediction[0][0] #'Diabetic' if prediction[0][0] > 0.5 else 'Non-Diabetic'
```

```python
# Example input data (replace with actual values)
sample_input = [5, 120, 80, 30, 40, 25.0, 0.5, 35]

# Call the function and print the result
result = predict_diabetes(sample_input)

print("Prediction:", result)
```
